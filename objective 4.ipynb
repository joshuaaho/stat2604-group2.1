```python
# Combine all indicies from each DataFrame
index = list(df.index) + list(df2.index) + list(df3.index)
```

```python
# Only use companies whose data are avaliable in all three years
unique, count = np.unique(index, return_counts=True)
unique = unique[count>2]
len(unique)
```

```python
# Extract those companies
df = df.loc[unique]
df2 = df2.loc[unique]
df3 = df3.loc[unique]
```

```python
# Combined all observations from the three years
combined_df = pd.concat([df,df2,df3])
combined_df

```

```python
#  Make instrument and year as the index i.e. Each observation can be identified by (year, instrument)

combined_df = combined_df.astype('float')
combined_df = combined_df.set_index(["Instrument","Year"])
table = pd.pivot_table(combined_df, index=["Instrument","Year"])
table


```

```python

corr = table.corr()
corr = corr.drop(["EScore * Board Diversity","GScore * Board Diversity","SScore * Board Diversity","Total Debt","Total Equity"], axis = 1)
corr = corr.drop(["EScore * Board Diversity","GScore * Board Diversity","SScore * Board Diversity","Total Debt","Total Equity"], axis = 0)

f, ax = plt.subplots(figsize=(12, 10))

# Generate a mask for upper traingle

mask = np.triu(np.ones_like(corr, dtype=bool))

# Configure a custom diverging colormap

cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap

sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```

```python
from linearmodels import PanelOLS
from linearmodels import RandomEffects

# Regression

y = table["ROA"]
table["Const"] = 1
x = ["Board Diversity","Board Size","EScore","EScore * Board Diversity","Firm Size","GScore","GScore * Board Diversity","Leverage Ratio","SScore","SScore * Board Diversity","Const"]
results = PanelOLS(y,table[x]).fit(cov_type='clustered',cluster_entity=True)
results
```
